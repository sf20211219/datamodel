```python
import matplotlib.pyplot as plt
import pandas as pd

# 총 매출과 목표 매출 데이터 입력
data = {
    'Month': ['Sep', 'Oct', 'Nov', 'Dec', 'Jan', 'Feb'],
    'Total Sales': [5280000, 5501000, 5469000, 5480000, 5533000, 5554000],
    'Target Sales': [5280000, 5500000, 5729000, 5968000, 6217000, 6476000]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))

# 총 매출 및 목표 매출 데이터 비교
plt.plot(df['Month'], df['Total Sales'], marker='o', label='Total Sales')
plt.plot(df['Month'], df['Target Sales'], marker='o', label='Target Sales')

plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.title('Total Sales vs Target Sales')
```

```python
# 총 매출, 광고 비용, 소셜 네트워크 비용 데이터 입력
data = {
    'Month': ['Sep', 'Oct', 'Nov', 'Dec', 'Jan', 'Feb'],
    'Total Sales': [5280000, 5501000, 5469000, 5480000, 5533000, 5554000],
    'Ad Costs': [1056000, 950400, 739200, 528000, 316800, 316800],
    'Social Media Costs': [0, 105600, 316800, 528000, 739200, 739200]
}

df = pd.DataFrame(data)

# 광고 및 소셜네트워크 비용의 총합 계산
total_ad_costs = df['Ad Costs'] + df['Social Media Costs']

# 비율 계산
ad_costs_ratio = (df['Ad Costs'] / df['Total Sales']) * 100
social_media_ratio = (df['Social Media Costs'] / df['Total Sales']) * 100

plt.figure(figsize=(10, 6))

plt.plot(df['Month'], ad_costs_ratio, marker='o', label='Ad Costs Ratio')
plt.plot(df['Month'], social_media_ratio, marker='o', label='Social Media Ratio')

plt.xlabel('Month')
plt.ylabel('Percentage (%)')
plt.title('Ad Costs and Social Media Costs Ratio to Total Sales')
```

```python
# 총 매출과 1온스별 단가 데이터 입력
data = {
    'Month': ['Sep', 'Oct', 'Nov', 'Dec', 'Jan', 'Feb'],
    'Unit Price per Ounce': [2.00, 2.00, 2.00, 1.90, 1.90, 1.90],
    'Total Sales': [5280000, 5501000, 5469000, 5480000, 5533000, 5554000],
}

df = pd.DataFrame(data)

# 각 월에 판매된 제품 수 계산
sold_units = df['Total Sales'] / df['Unit Price per Ounce']

plt.figure(figsize=(10, 6))

plt.plot(df['Month'], sold_units, marker='o', label='Sold Units')

plt.xlabel('Month')
plt.ylabel('Number of Units Sold')
plt.title('Number of Units Sold per Month')

plt.grid(True)
plt.legend()
plt.show()
```

```python
# 목표 매출 및 1온스별 단가 데이터 입력
data = {
    'Month': ['Sep', 'Oct', 'Nov', 'Dec', 'Jan', 'Feb'],
    'Unit Price per Ounce': [2.00, 2.00, 2.00, 1.90, 1.90, 1.90],
    'Target Sales': [5280000, 5500000, 5729000, 5968000, 6217000, 6476000],
}

df = pd.DataFrame(data)

# 각 월에 목표 매출을 달성하려면 몇 온스의 제품을 팔아야 하는지 계산
target_units = df['Target Sales'] / df['Unit Price per Ounce']

plt.figure(figsize=(10, 6))

# 각 월에 판매해야 하는 제품 수
plt.plot(df['Month'], target_units, marker='o', label='Target Units')

plt.xlabel('Month')
plt.ylabel('Number of Units to Achieve Target Sales')
plt.title('Number of Units to Achieve Target Sales per Month')

plt.grid(True)
plt.legend()
plt.show()
```

```python
# 총 매출과 1온스별 단가 데이터 입력
data_sales = {
    'Month': ['Sep', 'Oct', 'Nov', 'Dec', 'Jan', 'Feb'],
    'Unit Price per Ounce': [2.00, 2.00, 2.00, 1.90, 1.90, 1.90],
    'Total Sales': [5280000, 5501000, 5469000, 5480000, 5533000, 5554000],
}

df_sales = pd.DataFrame(data_sales)

# 각 월에 판매된 제품 수 계산
sold_units = df_sales['Total Sales'] / df_sales['Unit Price per Ounce']

# 목표 매출과 1온스별 단가 데이터 입력
data_target = {
    'Month': ['Sep', 'Oct', 'Nov', 'Dec', 'Jan', 'Feb'],
    'Unit Price per Ounce': [2.00, 2.00, 2.00, 1.90, 1.90, 1.90],
    'Target Sales': [5280000, 5500000, 5729000, 5968000, 6217000, 6476000],
}

df_target = pd.DataFrame(data_target)

# 각 월에 목표 매출을 달성하려면 몇 온스의 제품을 팔아야 하는지 계산
target_units = df_target['Target Sales'] / df_target['Unit Price per Ounce']

plt.figure(figsize=(15, 6))

# 총 매출 그래프
plt.subplot(1, 2, 1)
plt.plot(df_sales['Month'], sold_units, marker='o', label='Sold Units')
plt.xlabel('Month')
plt.ylabel('Number of Units Sold')
plt.title('Number of Units Sold per Month')
plt.grid(True)
plt.legend()

# 목표 매출 그래프
plt.subplot(1, 2, 2)
plt.plot(df_target['Month'], target_units, marker='o', label='Target Units', color='orange')
plt.xlabel('Month')
plt.ylabel('Number of Units to Achieve Target Sales')
plt.title('Number of Units to Achieve Target Sales per Month')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
